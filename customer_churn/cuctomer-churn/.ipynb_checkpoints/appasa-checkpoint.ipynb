#### Pairwise plot
Is it just many scatter plots stacked together or could it be much more?

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
customer_churn = pd.read_csv("churn_v2.csv")
customer_churn.head(10)
```

```python
customer_churn.tail(10)
```

```python
customer_churn.info()
```

```python
customer_churn.describe()
```

```python
customer_churnv2 = customer_churn.sample(n=1000, random_state=63)
```

#### Scatter plot

```python
sns.scatterplot(data=customer_churnv2, x='Age', y='Balance', hue='Exited')
plt.show()
```

```python
sns.scatterplot(data=customer_churnv2, x='CreditScore', y='EstimatedSalary', hue='Gender')
plt.show()
```

#### Pairwise

```python
sns.pairplot(customer_churnv2, vars=['CreditScore', 'Age', 'Balance', 'EstimatedSalary'], hue='Exited')
plt.show()
```

#### Features of a pair plot
1. Diagonal plots: Unvariate distribution
2. Off-Diagonal plots: Linear and non-linear relationships
3. Hue: For categorical vars

```python
sns.pairplot(customer_churnv2, vars=['Age', 'Tenure', 'Balance'], hue='Gender')
plt.show()
```

```python
sns.pairplot(customer_churnv2, vars=['NumOfProducts', 'EstimatedSalary', 'CreditScore'], hue='Exited')
plt.show()
```
